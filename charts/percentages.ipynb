```python
sns.set_theme(context="paper", font="serif", style="ticks", palette="tab10", font_scale=2.5)
matplotlib.rcParams['figure.figsize'] = [14, 7]

g = sns.catplot(
    data=df,
    kind='bar',
    x='Benchmark',
    y='Relative Difference (%)',
    hue='Metric',
    legend_out=False,
    aspect=2,
)
g.ax.axhline(color="black")
g.ax.set_xlabel("")

# Remove "Metric" from legend https://stackoverflow.com/a/51579663
handles, labels = g.ax.get_legend_handles_labels()
# I know it's supposed to be 1:, but this works. Don't ask me why
g.ax.legend(handles=handles[:], labels=labels[:])

savefig('percentages')
```
